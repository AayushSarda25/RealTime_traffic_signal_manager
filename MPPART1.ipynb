```python
import os
```

```python
model = YOLO("yolov8n.pt")

# Path to the folder containing images
folder_path = "/content/images/train"

# Vehicle class IDs in YOLO (car=2, motorcycle=3, bus=5, truck=7)
vehicle_classes = [2, 3, 5, 7]

# Check if the folder exists
if not os.path.exists(folder_path):
    print(f"Error: Folder '{folder_path}' does not exist.")
    exit()

# Iterate over all files in the folder
for filename in os.listdir(folder_path):
    image_path = os.path.join(folder_path, filename)

    # Check if it's an image file
    if filename.lower().endswith(('.jpg', '.jpeg', '.png', '.bmp')):
        try:
            # Read the image
            image = cv2.imread(image_path)
            if image is None:
                print(f"Skipping {filename}: Unable to read the image.")
                continue

            # Perform object detection
            results = model(image)

            # Count vehicles
            vehicle_count = 0
            for r in results:
                for box in r.boxes:
                    if int(box.cls) in vehicle_classes:
                        vehicle_count += 1

            print(f"Image: {filename} -> Vehicles detected: {vehicle_count}")

        except Exception as e:
            print(f"Error processing {filename}: {e}")
    else:
        print(f"Skipping {filename}: Not an image file.")
```

```python
import matplotlib.pyplot as plt

# Iterate over all files in the folder
for filename in os.listdir(folder_path):
    image_path = os.path.join(folder_path, filename)

    # Check if it's an image file
    if filename.lower().endswith(('.jpg', '.jpeg', '.png', '.bmp')):
        try:
            # Read the image
            image = cv2.imread(image_path)
            if image is None:
                print(f"Skipping {filename}: Unable to read the image.")
                continue

            # Perform object detection
            results = model(image)

            # Count vehicles
            vehicle_count = 0
            for r in results:
                for box in r.boxes:
                    if int(box.cls) in vehicle_classes:
                        vehicle_count += 1

            # Display the image and vehicle count
            plt.figure(figsize=(8, 6))
            plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB for display
            plt.title(f"Image: {filename}\nVehicles detected: {vehicle_count}")
            plt.axis('off')  # Hide axis ticks and labels
            plt.show()

        except Exception as e:
            print(f"Error processing {filename}: {e}")
    else:
        print(f"Skipping {filename}: Not an image file.")
```
